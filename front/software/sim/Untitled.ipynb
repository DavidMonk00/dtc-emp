```python
import numpy as np
import pandas as pd
```

```python
def parse(string):
    data = []
    i = 0
    start, end, depth = 0, 0, 0
    while (i < len(string)):
        if (string[i] == '{'):
            if (depth == 0):
                start = i
            depth += 1
        elif (string[i] == '}'):
            depth -= 1
            if (depth == 0):
                end = i
                data.append(string[start+1:end])
        i += 1
    if len(data) == 0:
        return string
    else:
        for i in range(len(data)):
            data[i] = parse(data[i])
    return data
```

```python
lines = [line.strip() for line in open("list2.lst")]
for i in range(2, len(lines)):
     lines[i] = parse(lines[i])
```
